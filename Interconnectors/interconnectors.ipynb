```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("InterconnectorFlows_18102023_to_181024_30min.csv")
```

```python
df = df.set_index("StartTime") 
df.sort_values(["SettlementDate", "SettlementPeriod"], inplace=True) # Ordering by date then time
df.head()
```

```python
df = df[df["FuelType"] != 'INTGRNL'] # removing greenlink as all generation is 0
df.head()
```

```python
df[df["FuelType"] == 'INTNEM'].head()
```

```python
inters = set(df["FuelType"])
inters_info = {i:[0]*4 for i in set(df["FuelType"])}

for inter in inters:
    inter_info = df[df["FuelType"] == inter]["Generation"]
    inters_info[inter][0] = inter_info.sum()
    inters_info[inter][1] = inter_info.max()
    inters_info[inter][2] = inter_info.min()
    inters_info[inter][3] = len(inter_info)
    
```
